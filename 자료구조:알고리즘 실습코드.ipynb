### 2강. 선형 배열(Linear Array)
**[실습1] 정렬된 리스트에 주어진 원소 삽입하기**<br>
: 리스트 L 과 정수 x 가 인자로 주어질 때, 리스트 내의 올바른 위치에 x 를 삽입하여 그 결과 리스트를 반환하는 함수 solution() 완성

1. 주어진 원소를 삽입할 위치 찾기
2. 해당 위치에 원소 삽입, 결과는 여전히 정렬된 상태를 유지하는 리스트 형태

```python
def solution(L, x):
    i = 0
    
    while x > L[i]: #인덱스 0에서부터 시작해서 대소관계 비교
        i += 1
        if i >= len(L):
            break
            
    L.insert(i, x)
    return L
```

### 2강. 선형 배열(Linear Array)
**[실습2] 주어진 리스트에서 특정 원소 찾기**<br>
: 인자로 주어지는 리스트 L 내에서, 또한 인자로 주어지는 원소 x 가 발견되는 모든 인덱스를 구하여 이 인덱스들로 이루어진 리스트를 반환하는 함수 solution() 완성

1. index() 메서드 이용 (있는 원소 추출)
2. 리스트 슬라이싱을 이용해서 반복적으로 찾기 (여러번 들어있는 경우까지 고려)

```python
def solution(L, x):    
    answer = []
    for i in range(0, len(L)): #i=index
        if L[i] == x:
            answer.append(i)
    if len(answer) == 0:
        answer.append(-1)
    return answer
```

### 3강. 정렬(Sort), 탐색(Search)
**[실습] 이진 탐색 구현해보기**<br>
: 리스트 L 과, 그 안에서 찾으려 하는 원소 x 가 인자로 주어질 때, x 와 같은 값을 가지는 원소의 인덱스를 리턴하는 함수 solution() 완성

```python
def solution(L, x):
    lower = 0
    upper = len(L) -1
    idx = -1
    
    while lower <= upper:
        middle = (lower+upper)//2
        
        if L[middle] == x:
            return middle
        elif L[middle] < x:
            lower = middle+1
        else:
            upper = middle-1
    return idx
```

### 4강. 재귀함수 (Recursive function)
**[실습] 피보나치 수열 구현하기**<br>
: 인자로 0 또는 양의 정수인 x 가 주어질 때, Fibonacci 순열의 해당 값을 구하여 반환하는 함수 solution() 완성

- F0 = 0
- F1 = 1
- Fn = Fn - 1 + Fn - 2, n >= 2



```python
def solution(x):
    if x < 2: #0과 1일 경우 자기 자신 그대로 호출
        return x
    
    f1 = 0
    f2 = 1
    for i in range(2, x+1):
        f = f1 + f2
        f1, f2 = f2, f
    return f
```

### 5강. 재귀함수 (Recursive function)
**[실습] 재귀적 이진 탐색 구현하기**<br>
: 리스트 L 과, 그 안에서 찾으려 하는 원소 x 가 인자로 주어지고, 또한 탐색의 대상이 되는 리스트 내에서의 범위 인덱스가 l 부터 u 까지로 (인자로) 정해질 때, x 와 같은 값을 가지는 원소의 인덱스를 리턴하는 함수 solution() 완성

```python
def binary_search(L, x, lower, upper):
  if lower > upper:
    return -1
  middle = (lower+upper)//2

  if x == L[middle]:
    return middle
  elif x < L[middle]:
    return binary_search(L, x, lower, middle-1)
  else:
    return binary_search(L, x, middle+1, upper)
```

### 7강. 연결 리스트 (Linked Lists)
**[실습] 연결 리스트 순회 구현하기**<br>
: 추상적 자료구조로 LinkedList 라는 이름의 클래스가 정의되어 있다고 가정하고, 이 리스트를 처음부터 끝까지 순회하는 메서드 traverse() 완성

```python
class Node:
    def __init__(self, item):
        self.data = item
        self.next = None

class LinkedList:
    def __init__(self):
        self.nodeCount = 0
        self.head = None
        self.tail = None

    def getAt(self, pos):
        if pos < 1 or pos > self.nodeCount:
            return None
        i = 1
        curr = self.head
        while i < pos:
            curr = curr.next
            i += 1
        return curr

    #리스트 순회하는 traverse() 함수
    def traverse(self):
        linkedlist = []
        curr = self.head
        while curr != None:
            linkedlist.append(curr.data)
            curr = curr.next
        return linkedlist
```

### 8강. 연결 리스트 (Linked Lists)
**[실습] 연결 리스트 노드 삭제하기**<br>
: popAt() 함수 완성

```python
def popAt(self, pos):
    if pos < 1 or pos > self.nodeCount:
        raise IndexError
        
    curr = self.getAt(pos)
        
    #삭제하려는 노드가 맨 앞인 경우
    if pos == 1:
        if self.nodeCount == 1: #유일한 노드인 경우
            self.head = None
            self.tail = None
        else:
            self.head = curr.next
        
    #삭제하려는 노드가 맨 뒤인 경우
    elif pos == self.nodeCount:
        prev = self.getAt(pos-1)
        prev.next = None
        self.tail = prev
        
    #삭제하려는 노드가 중간에 오는 경우
    else:
        prev = self.getAt(pos-1)
        prev.next = curr.next
        
    self.nodeCount -= 1
    return curr.data #삭제한 데이터 return
```

### 11강. 스택 (Stack)
**[실습] 스택을 이용한 수식의 괄호 검사**<br>
: 소괄호: ( )
중괄호: { }
대괄호: [ ]
를 포함할 수 있는 수식을 표현한 문자열 expr 이 인자로 주어질 때, 이 수식의 괄호가 올바르게 여닫혀 있는지를 판단하는 함수 solution() 완성

```python
class ArrayStack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]


def solution(expr):
    match = {
        ')': '(',
        '}': '{',
        ']': '['
    }
    S = ArrayStack()
    for c in expr:
        if c in '({[':
            S.push(c)
        elif c in match:
            if S.isEmpty():
                return False
            else:
                t = match[c]
                if t != S.pop() :
                    return False
    return True
```

### 12강. 스택의 응용: 수식 후위 표기법
**[실습] 중위표현 수식 > 후위표현 수식**<br>
: 중위 표기법을 따르는 수식 S 가 인자로 주어질 때, 이 수식을 후위 표기법을 따르는 수식으로 변환하여 반환하는 함수 solution() 완성
- 피연산자 → 그냥 출력 (A, B 등)
- ( → 스택에 push
- ) → ( 이 나올 때까지 스택에서 pop, 출력
- 연산자 → 스택에서 이보다 높거나 같은 우선순위의 것들을 pop, 출력
(맨 위의 연산자와의 우선순위를 비교하기 위해 스택의 **`peek()`** 연산 이용)
- 우선순위가 가장 낮은 연산자는 스택에 push
- 중위 표현식의 끝을 만나면 스택에 남아 있는 연산자는 모두 pop, 출력
→ 스택에 남아 있는 연산자 모두를 pop()할 때 사용하는 순환문
`while not opStack.isEmpty():`

```python
class ArrayStack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]

def solution(S):
    prec = { #숫자 작을수록 우선순위 낮음
    '*': 3, '/': 3,
    '+': 2, '-': 2,
    '(': 1
    }
    opStack = ArrayStack()
    answer = ''
    for c in S:
        if c == '(' : #여는 괄호인 경우
            opStack.push(c)
        
        elif c == ')': #닫는 괄호인 경우
            while opStack.peek() != '(':
                answer += opStack.pop()
            opStack.pop()
        
        else:
            if c in prec: #연산자인 경우 우선순위 비교
                if opStack.isEmpty():
                    opStack.push(c)
                elif prec[opStack.peek()] >= prec[c]: #스택에 있는 것이 우선순위가 높거나 같다면
                    while not opStack.isEmpty() and prec[opStack.peek()] >= prec[c]:
                        answer += opStack.pop() #스택에 있는 것 pop
                    opStack.push(c)
                else: #지금의 것이 우선순위가 더 높다면 스택에 넣기
                    opStack.push(c)
            else: #피연산자인 경우
                answer += c 
    while not opStack.isEmpty(): #중위표현식의 끝인 경우
        answer += opStack.pop()
    return answer
```

### 13강. 스택의 응용: 후위 표기 수식 계산
**[실습] 후위표현 수식 계산**<br>

- postfixEval() - 이번 강의의 연습문제입니다. 함수의 내용을 완성하세요.<br>

즉, 두 개의 함수 infixToPostfix() 와 postfixEval() 을 구현하는 연습입니다. 스택을 이용하기 위하여 class ArrayStack 이 정의되어 있으므로 그것을 활용하세요.

```python
class ArrayStack:

    def __init__(self):
        self.data = []

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

    def push(self, item):
        self.data.append(item)

    def pop(self):
        return self.data.pop()

    def peek(self):
        return self.data[-1]


def splitTokens(exprStr):
    tokens = []
    val = 0
    valProcessing = False
    for c in exprStr:
        if c == ' ':
            continue
        if c in '0123456789':
            val = val * 10 + int(c)
            valProcessing = True
        else:
            if valProcessing:
                tokens.append(val)
                val = 0
            valProcessing = False
            tokens.append(c)
    if valProcessing:
        tokens.append(val)

    return tokens


#앞에서는 그냥 '' 원소 형태, 이번에는 빈 리스트 형태
def infixToPostfix(tokenList):
    prec = { #숫자 작을수록 우선순위 낮음
    '*': 3, '/': 3,
    '+': 2, '-': 2,
    '(': 1
    }
    opStack = ArrayStack()
    postfixList = []
    for c in tokenList:
        if c == '(' : #여는 괄호인 경우
            opStack.push(c)
        
        elif c == ')': #닫는 괄호인 경우
            while opStack.peek() != '(':
                postfixList += opStack.pop()
            opStack.pop()
        
        else:
            if c in prec: #연산자인 경우 우선순위 비교
                if opStack.isEmpty():
                    opStack.push(c)
                elif prec[opStack.peek()] >= prec[c]: #스택에 있는 것이 우선순위가 높거나 같다면
                    while not opStack.isEmpty() and prec[opStack.peek()] >= prec[c]:
                        postfixList += opStack.pop() #스택에 있는 것 pop
                    opStack.push(c)
                else: #지금의 것이 우선순위가 더 높다면 스택에 넣기
                    opStack.push(c)
            else: #피연산자인 경우
                postfixList.append(c) 
    while not opStack.isEmpty(): #중위표현식의 끝인 경우
        postfixList += opStack.pop()
    return postfixList


def postfixEval(tokenList):
    anStack = ArrayStack()
    for c in tokenList:
        if c == '+':
            first_popdata = anStack.pop()
            second_popdata = anStack.pop()
            temp = first_popdata + second_popdata
            anStack.push(temp)
        elif c == '-':
            first_popdata = anStack.pop()
            second_popdata = anStack.pop()
            temp = second_popdata - first_popdata #빼기는 순서 바뀐거에 유의
            anStack.push(temp)
        elif c == '/':
            first_popdata = anStack.pop()
            second_popdata = anStack.pop()
            temp = second_popdata / first_popdata #나누기는 순서 바뀐거에 유의
            anStack.push(temp)
        elif c == '*':
            first_popdata = anStack.pop()
            second_popdata = anStack.pop()
            temp = first_popdata * second_popdata
            anStack.push(temp)
        else: #피연산자인 경우
            anStack.push(c)
    answer = anStack.pop()
    return answer


def solution(expr):
    tokens = splitTokens(expr)
    postfix = infixToPostfix(tokens)
    val = postfixEval(postfix)
    return val
```

### 21강. 이진 탐색 트리 (Binary Search Trees)
**[실습]이진 탐색 트리에서 노드의 삭제 연산 구현**<br>
: 초기 코드에 주어진 class Node 와 class BinSearchTree 를 기반으로, 이진 탐색 트리 (binary search tree) 에서 지정된 원소를 삭제하는 remove(key) 연산의 구현을 완성

```python
class Node:

    def __init__(self, key, data):
        self.key = key
        self.data = data
        self.left = None
        self.right = None


    def insert(self, key, data):
        if key < self.key:
            if self.left:
                self.left.insert(key, data)
            else:
                self.left = Node(key, data)
        elif key > self.key:
            if self.right:
                self.right.insert(key, data)
            else:
                self.right = Node(key, data)
        else:
            raise KeyError('Key %s already exists.' % key)


    def lookup(self, key, parent=None):
        if key < self.key:
            if self.left:
                return self.left.lookup(key, self)
            else:
                return None, None
        elif key > self.key:
            if self.right:
                return self.right.lookup(key, self)
            else:
                return None, None
        else:
            return self, parent


    def inorder(self):
        traversal = []
        if self.left:
            traversal += self.left.inorder()
        traversal.append(self)
        if self.right:
            traversal += self.right.inorder()
        return traversal


    def countChildren(self):
        count = 0
        if self.left:
            count += 1
        if self.right:
            count += 1
        return count


class BinSearchTree:

    def __init__(self):
        self.root = None


    def insert(self, key, data):
        if self.root:
            self.root.insert(key, data)
        else:
            self.root = Node(key, data)


    def lookup(self, key):
        if self.root:
            return self.root.lookup(key)
        else:
            return None, None


    def remove(self, key):
        node, parent = self.lookup(key)
        if node:
            nChildren = node.countChildren()
            # 1. The simplest case of no children (말단 리프 노드인 경우)
            if nChildren == 0:
                # 만약 parent 가 있으면
                # node 가 왼쪽 자식인지 오른쪽 자식인지 판단하여
                # parent.left 또는 parent.right 를 None 으로 하여
                # leaf node 였던 자식을 트리에서 끊어내어 없앱니다.
                if parent:
                    if parent.key > node.key:
                        parent.left = None
                    else:
                        parent.right = None
                # 만약 parent 가 없으면 (node 는 root 인 경우)
                # self.root 를 None 으로 하여 빈 트리로 만듭니다.
                else:
                    self.root = None
            
            # 2. When the node has only one child (자식이 하나인 경우)
            elif nChildren == 1:
                # 하나 있는 자식이 왼쪽인지 오른쪽인지를 판단하여
                # 그 자식을 어떤 변수가 가리키도록 합니다.
                if node.left:
                    x = node.left #삭제하는 노드=x
                else:
                    x = node.right
                # 만약 parent 가 있으면
                # node 가 왼쪽 자식인지 오른쪽 자식인지 판단하여
                # 위에서 가리킨 자식을 대신 node 의 자리에 넣습니다.
                if parent:
                    if parent.left  == node:
                        parent.left = x
                    else:
                        parent.right = x
                # 만약 parent 가 없으면 (node 는 root 인 경우)
                # self.root 에 위에서 가리킨 자식을 대신 넣습니다.
                else:
                    self.root = x
            
            # 3. When the node has both left and right children (자식이 둘인 경우)
            else:
                parent = node
                successor = node.right #그 다음으로 큰 노드
                # parent 는 node 를 가리키고 있고,
                # successor 는 node 의 오른쪽 자식을 가리키고 있으므로
                # successor 로부터 왼쪽 자식의 링크를 반복하여 따라감으로써
                # 순환문이 종료할 때 successor 는 바로 다음 키를 가진 노드를,
                # 그리고 parent 는 그 노드의 부모 노드를 가리키도록 찾아냅니다.
                while successor.left:
                    parent = successor
                    successor = successor.left
  
                # 삭제하려는 노드인 node 에 successor 의 key 와 data 를 대입합니다.
                node.key = successor.key
                node.data = successor.data

                # 이제, successor 가 parent 의 왼쪽 자식인지 오른쪽 자식인지를 판단하여
                # 그에 따라 parent.left 또는 parent.right 를
                # successor 가 가지고 있던 (없을 수도 있지만) 자식을 가리키도록 합니다.
                if parent.left == successor:
                    parent.left = successor.right
                else:
                    parent.right = successor.right

            return True

        else:
            return False


    def inorder(self):
        if self.root:
            return self.root.inorder()
        else:
            return []


def solution(x):
    return 0
```

### 23강. 힙 (Heaps)
